### 1.4 Extract Frames
I then extracted the frames out for labeling.
I used python cv2 library to extract video frames as images every 15 frames interval.

```python
import cv2
import os

# Load the video
video_path = '../datasets/videos/train.mp4'
output_folder = '../datasets/images/'
os.makedirs(output_folder, exist_ok=True)

cap = cv2.VideoCapture(video_path)
frame_rate = 15  # the interval to save frames

frame_count = 0
saved_frame_count = 0

while True:
    success, frame = cap.read()
    if not success:
        break  # no more frames, or can't read video

    # Save frame every 15 frames
    if frame_count % frame_rate == 0:
        output_path = os.path.join(output_folder, f"frame_{saved_frame_count}.jpg")
        cv2.imwrite(output_path, frame)
        saved_frame_count += 1

    frame_count += 1

cap.release()
print(f"Extracted {saved_frame_count} images.")
```

### 1.5 Filter Out Unwanted Images

### 1.6 Initial Labeling Using Grounding Dino

I use zero-shot model, Grounding Dino, for initial labeling.  
I use simple text prompts such as colors, shapes and size. For example instead of directional sign board, I use green rectangle board. I also can adjust the confidence threshold, so grounding dino can detect more objects.

```python
# Labeled dataset is around 240mb

from roboflow import Roboflow

rf = Roboflow(api_key="Insert API Key")
project = rf.workspace("roadobject-nvbbi").project("singapore_road_objects_v2")
version = project.version(3)
dataset = version.download("yolov8")
```

```python
import glob
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# path to the labels folders
label_folder_paths = ['../datasets/singapore_road_objects_v2.v3i.yolov8/train/labels/',
                '../datasets/singapore_road_objects_v2.v3i.yolov8/valid/labels/',
                '../datasets/singapore_road_objects_v2.v3i.yolov8/test/labels/'
               ]
```

```python
# class count for the entire dataset
total_no_class = []

# class number for each frame
class_in_frames = []

for folder_path in label_folder_paths:
    for file_path in glob.glob(f'{folder_path}*.txt'):
        with open(file_path, 'r', encoding='utf-8') as file:
            # split text content by line and get the class number for every line
            lines = file.read().splitlines()
            class_in_frame = [line.split()[0] for line in lines]
            
            # combine all class number into a list
            total_no_class.extend(class_in_frame)
            
            # add class number in each frame to a list
            class_in_frames.append(class_in_frame)
            
```

```python
# get class distribution for each frame
no_class_in_frames = []

for class_in_frame in class_in_frames:
    no_class_in_frames.append(len(class_in_frame))
```

```python
plt.figure(figsize=(10,6))
sns.histplot(no_class_in_frames, bins=range(1,24))
plt.xticks(range(1,24))

# Calculate mean and median
mean_value = np.mean(no_class_in_frames)
median_value = np.median(no_class_in_frames)

# Add mean and median lines
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean_value:.2f}')
plt.axvline(median_value, color='green', linestyle='dashed', linewidth=1.5, label=f'Median: {median_value:.2f}')

# Add labels and title
plt.xlabel('Class Count', fontsize=14)
plt.ylabel('Number of Frames', fontsize=14)
plt.title('Class Count Per Frame Histogram', fontsize=16, fontweight='bold')

plt.legend()
plt.show();

print(f'Class count per frame mean value is {mean_value}')
print(f'Class count per frame median value is {median_value}')
```

Class count per frame distribution is normal where mean value and median value is almost the same.
Every image roughly has 9 objects / class.  

The number of objects per frame is reasonable, allowing the model to run inference in real time.
